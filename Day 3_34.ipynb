```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([3,13,19,31,38,45,58,65,71,88])
y = np.array([11,16,24,37,41,58,56,78,81,95])
z = np.array([81,80,79,63,54,32,29,26,18,14])
```

```python
plt.title('My first plot')
plt.xlabel('First value')
plt.xlabel('Second value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))

plt.plot(x,z)
```

```python
plt.title('My first plot')
plt.xlabel('First value')
plt.xlabel('Second value')
plt.grid()
plt.plot(x,y,color='red',marker='o',ls=':')
```

```python
plt.title('My first plot')
plt.xlabel('First value')
plt.xlabel('Second value')
plt.grid()
plt.plot(x,y,color='red',label = '2023')
plt.plot(x,z,color='green',label = '2022')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First value')
plt.xlabel('Second value')
plt.grid()
plt.plot(x,z,lw=5,marker='o',ms=20,mfc='y')
plt.savefig('myplot.png')
```

### Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
```

 ### separate the input and output data

```python
x=df['YearsExperience'].values
y=df['Salary']
```

```python
x
```

```python
y;
```

```python
x=x.reshape(30,1)
```

```python
x
```

### import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

### create the object of linear regression

```python
regressor=LinearRegression()
```

### train the algorithm

```python
regressor.fit(x,y)
```

### predict on nthe unseen data

```python
regressor.predict([[15]])
```

```python
regressor.predict([[20]])
```

```python
regressor.score(x,y)
```
